### READ WORKBOOK FROM EXCEL FILE

```python
kpmg_wb = pd.read_excel(io=sf['KPMG_VI_New_raw_data_update_final - Copy.xlsx'], sheet_name=None)
kpmg_wb.keys()
```

#### TRANSACTIONS DATABASE

```python
trans = kpmg_wb['Transactions']
trans.head()
```

#### CLEANING TRANSACTIONS DATAFRAME

```python
trans = trans.dropna(thresh=1, axis=1).dropna(thresh=1, axis=0)
trans.head()
```

```python
trans = trans.set_axis(labels=trans.loc[1], axis=1).drop(labels=1, axis=0)
trans.head()
```

```python
trans.reset_index(drop=True).head()

# Alternatively:
# trans.reset_index().drop(labels='index', axis=1)
```

```python
trans = trans.reset_index(drop=True).rename_axis(axis=1, mapper='')
```

```python
trans
```

```python
trans.info()
```

#### ADDITIONAL TRANSACTIONS FOR OCT-NOV-DEC

```python
trans2 = kpmg_wb['Oct-Nov-Dec_Trans']
trans2
```

```python
# REMOVE INVALID COLUMNS AND ROWS FROM DATAFRAME
trans2 = trans2.dropna(thresh=1, axis=1).dropna(thresh=1, axis=0)
trans2
```

```python
# Use first row as new header
# then remove the first row
trans2 = trans2.set_axis(labels=trans2.loc[0], axis=1).drop(labels=0, axis=0)
trans2.head()
```

```python
# reset to pandas implicit index
trans2 = trans2.reset_index().drop(labels='index', axis=1)
trans2
```

```python
trans2.info()
```

#### Combine oct-nov-dec transactions to previous transactions

```python
all_trans = pd.concat([trans, trans2], ignore_index=True).sort_values('transaction_id').reset_index().drop('index', axis=1)
all_trans.head()
```

```python
all_trans.info()
```

#### Remove duplicate transactions

```python
all_trans.loc[all_trans['transaction_id'].duplicated()].head()
```

```python
# select rows where transaction_id is unique
all_trans = all_trans.loc[~(all_trans['transaction_id'].duplicated() )].reset_index().drop('index', axis=1)
all_trans.head()
```

### Customer Demographic dataframe

```python
cust_demog = kpmg_wb['CustomerDemographic']
cust_demog.head()
```

```python
# remove completely null columns and rows
cust_demog = cust_demog.dropna(thresh=1, axis=1).dropna(thresh=2, axis=0)
cust_demog.head()
```

```python
# make the entire first row our header
# then remove the first row after header has been set
cust_demog = cust_demog.set_axis(cust_demog.loc[1], axis=1).drop(labels=1, axis=0)
cust_demog.head()
```

```python
# reset index to default pandas implicit index
# then remove the former index
cust_demog = cust_demog.reset_index().drop(labels=['index'], axis=1)
cust_demog.head()
```

```python
cust_demog.info()
```

#### Customer Address dataframe

```python
cust_addr = kpmg_wb['CustomerAddress']
cust_addr.head()
```

```python
# remove invalid columns and rows
cust_addr = cust_addr.dropna(thresh=1, axis=1).dropna(thresh=2, axis=0)
cust_addr
```

```python
# adopt first row as new header
# then remove that row
cust_addr = cust_addr.set_axis(cust_addr.loc[1], axis=1).drop(labels=1, axis=0)
cust_addr
```

```python
# reset the index to pandas implicit index
# then remove the old index column
cust_addr = cust_addr.reset_index().drop(labels='index', axis=1)
cust_addr
```

```python
# change the name of column axis from 1 to ''
cust_addr = cust_addr.rename_axis(axis=1, mapper='')
cust_addr
```

```python
cust_addr.info()
```

### Combine Customer Demographic and Customer Address dataframes

```python
custs = pd.merge(left=cust_demog, right=cust_addr, how='left', on='customer_id').sort_values('customer_id')
custs
```
